```python
# Importing Libraries
import ast
import calendar
import seaborn as sns
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt 

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)  
```

```python
# Filter Data for Data Anlyst Roles in Germany
df_DA_GE = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'Germany')].copy() 
df_DA_GE
```

```python
#Add a column: Job Posted Month
df_DA_GE['job_posted_mo'] = df_DA_GE['job_posted_date'].dt.month
df_DA_GE
```

```python
#Explode Method to get each skill of skills list in a single row
df_skills = df_DA_GE.explode('job_skills')
```

```python
# Group by month and job_skills
df_DA_GE_Pivot = df_skills.pivot_table(index='job_posted_mo', columns='job_skills', aggfunc='size', fill_value=0)
df_DA_GE_Pivot
```

```python
# Sorts by count
df_DA_GE_Pivot.loc['Total'] = df_DA_GE_Pivot.sum()
df_DA_GE_Pivot = df_DA_GE_Pivot[df_DA_GE_Pivot.loc['Total'].sort_values(ascending=False).index]
df_DA_GE_Pivot = df_DA_GE_Pivot.drop('Total')
df_DA_GE_Pivot
```

```python
# changes month number to month name
df_DA_GE_Pivot.index = df_DA_GE_Pivot.index.map(lambda x: calendar.month_abbr[int(x)])
```

```python
# Plot Line Chart of the Top 5 Skills for Data Analysts in Germany
df_DA_GE_Pivot.iloc[:, :5].plot(kind='line')

sns.set_theme(style='ticks')
sns.despine() # remove top and right spines
plt.title('Trending Top Skills for Data Analysts in Germany')
plt.ylabel('Count')
plt.xlabel('2023')
plt.ylim(0, 500)

plt.show()
```
